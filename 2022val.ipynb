```python
import pandas as pd
import matplotlib.pyplot as plt

# Завантаження даних
df = pd.read_csv("exchange.csv")

# Перевірка структури таблиці
print(df.head())

# Фільтрація даних для отримання валют і курсу
df = df[['cc', 'rate']]  # Оставляємо тільки стовпці з кодом валюти і курсом

# Перетворення стовпця 'rate' на числові значення (якщо вони є текстовими)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Видалення рядків з пропущеними значеннями в стовпці 'rate'
df = df.dropna(subset=['rate'])

# Побудова графіка
plt.figure(figsize=(30, 20))  # Розмір графіка
bars = plt.bar(df['cc'], df['rate'], color='skyblue')  # Створюємо стовпчиковий графік

# Додавання числових позначок на стовпці з налаштуваннями
for bar in bars:
    yval = bar.get_height()
    
    # Масштабування відстані для кращої видимості
    if yval > 10:
        offset = 0.5  # Для більших значень зміщення вище
    else:
        offset = 0.1  # Для менших значень зміщення менше

    plt.text(bar.get_x() + bar.get_width()/2, yval + offset, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

# Форматування графіка
plt.xticks(rotation=45)
plt.title('Курс валют 06.03.2025')
plt.xlabel('Валюта')
plt.ylabel('Курс')

# Відображення графіка
plt.tight_layout()  # Покращує виведення графіка, щоб не обрізалися мітки
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Завантаження даних
df = pd.read_csv("exchange.csv")

# Перевірка структури таблиці
print(df.head())

# Фільтрація даних для отримання валют і курсу
df = df[['cc', 'rate']]  # Оставляємо тільки стовпці з кодом валюти і курсом

# Перетворення стовпця 'rate' на числові значення (якщо вони є текстовими)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Видалення рядків з пропущеними значеннями в стовпці 'rate'
df = df.dropna(subset=['rate'])

# Побудова графіка
plt.figure(figsize=(20, 10))  # Розмір графіка
bars = plt.bar(df['cc'], df['rate'], color='skyblue')  # Створюємо стовпчиковий графік

# Логарифмічна шкала для осі Y
plt.yscale('log')

# Додавання числових позначок на стовпці з налаштуваннями
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

# Форматування графіка
plt.xticks(rotation=45)
plt.title('Курс валют 06.03.2025')
plt.xlabel('Валюта')
plt.ylabel('Курс (логарифмічна шкала)')

# Відображення графіка
plt.tight_layout()  # Покращує виведення графіка, щоб не обрізалися мітки
plt.show()
```
